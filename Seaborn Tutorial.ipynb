# Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
data.corr()
```

```python
sns.heatmap(data.corr())
```

# Joinplot
A joinplot allows the study the relationship between 2 numeric variables. The central char display their correlation. It usually a scatterplot a hexbin plot, a 2D histogram or a 2D density plot
## Bivariate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="reg")
```

# Pair plot
A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(data)
```

```python
data.smoker.value_counts()
```

```python
sns.pairplot(data,hue="smoker")
```

# Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(data["tip"])
# If kde are not presant then it will give probability density function.(exm-0.0,0.1 persantage)
```

```python
sns.distplot(data["tip"],kde=False,bins=10)
```

# Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
sns.countplot("sex",data=data)
#count plot
```

```python
sns.countplot("smoker",data=data)
```

```python
sns.countplot("day",data=data)
```

```python
sns.countplot(y="sex",data=data)
```

```python
 ## Bar plot
sns.barplot(x='sex',y='total_bill',data=data)
```

# Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot("sex","total_bill",data=data)
```

```python
sns.boxplot(x="day", y="total_bill", data=data,palette='rainbow')
```

```python
sns.boxplot(x="day", y="total_bill", data=data)
#default colour if we not define
```

```python
sns.catplot(x="sex", y="total_bill", hue="smoker",
            kind="violin", data=data, palette = 'magma')
```
